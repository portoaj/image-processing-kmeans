```
from kmeans import KMeans
#Create an instance of the kmeans class with 3 dimensions (3 dimensions is required for image processing)
km = KMeans(3)
```

```
#Use the elbow method to find the amount of clusters in the image and save the inertias to be graphed later
km.k, km.inertias = km.elbow(15, 80, "test.bmp")
#Now that the k (# of clusters) is set train the actual model
km.train(80)
```

```
#Graph the different inertias
print(km.inertias)
km.graphInertias()
```

```
#Graph the points and clusters in RGB (x,y,z) space
km.outputGraph()
```

```
#Render an image showing each points nearest centroid by its centroid's RGB value
#km.outputImage() This gives a better output but doesn't work in google colab
km.outputImageAlternate()
```
